```
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt


t= np.linspace(0,6,400)
plt.plot(t,1+ (signal.sawtooth(np.pi*1*t,0.5)))
plt.axhline(y=0, color='r', linestyle='-', label = 'lr_min')
plt.axhline(y=2, color='y', linestyle='-', label = 'lr_max')
plt.title("Triangular Schedule")
plt.show()
```
